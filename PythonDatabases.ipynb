```python
import psycopg2 as ps
import pandas as pd
import numpy as np
```

```python
myDataFrame = pd.read_csv('500-us-users.csv')
myDataFrame
```

```python
myDataFrame1 = myDataFrame.T.head(7).T
myDataFrame1
```

```python
myDataFrame2 = myDataFrame.T.tail(2).T
myDataFrame2
```

```python
connection = ps.connect(database="postgres", user="postgres", password="postgres", host="127.0.0.1", port="5432")

query = connection.cursor()

query.execute("create schema if not exists usa_users")

query.execute("SET search_path TO usa_users")
```

```python
phoneTable = "phone"
usersTable = "userdetails"

createPhoneTable = "create table "+phoneTable+ "(phone_id numeric(5,0) not null primary key, phone1 varchar(50) not null, phone2 varchar(50) not null);"
query.execute(createPhoneTable)
```

```python
createUsersTable = "create table "+usersTable+ "(user_id numeric(5,0) not null primary key, first_name varchar(50) not null, last_name varchar(50) not null, address varchar(100) not null,city varchar(50) not null, county varchar(20) not null, state varchar(50) not null, zip varchar(10) not null,phone_id numeric(5,0) not null);"
query.execute(createUsersTable)
```

```python
for index, row in myDataFrame1.iterrows():
    insertdata = "('"+str(index)+ "', '"+row[0]+"','"+str(row[1])+"','"+str(row[2])+"','"+str(row[3])+"','"+str(row[4])+"','"+str(row[5])+"','"+str(row[6])+"', '"+str(index)+ "')"
    
    query.execute("INSERT INTO userdetails values "+insertdata)
    print( "row inserted:", insertdata)
    
   
```
